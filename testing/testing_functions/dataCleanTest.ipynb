```python
tradesDf = pd.read_csv("/home/jake/TAQ-Query-Scripts/data/raw_data/2020-01-02/AAPL_trades.csv")
quotesDf = pd.read_csv("/home/jake/TAQ-Query-Scripts/data/raw_data/2020-01-02/AAPL_quotes.csv") 
# get data
allData = pd.concat([tradesDf, quotesDf], axis=0)

# intialize preprocessing pipeline
process_pipeline = make_pipeline(Data_Preprocessor())

# clean data
df_clean = process_pipeline.fit_transform(allData)

df_clean.head(20)
```

```python

df_clean2 = df_clean[((df_clean['National_BBO_Indicator']==4) | (pd.isna(df_clean['National_BBO_Indicator']))
                      & ((df_clean['Sale_Condition']=='@   ') | (pd.isna(df_clean['Sale_Condition'])))) & ((df_clean['Trade_Correction_Indicator']==0) | (pd.isna(df_clean['Trade_Correction_Indicator'])))]
df_clean2.index=pd.to_datetime(df_clean2.index)
fig,ax=plt.subplots(1,1,figsize=(14,10))
ax.scatter(df_clean2.index,df_clean2['Bid_Price'],s=10,label='Bid Price',color='blue',marker='o')
ax.scatter(df_clean2.index,df_clean2['Offer_Price'],s=10,label='Offer Price',color='red',marker='o')
ax.scatter(df_clean2.index,df_clean2['Trade_Price'],s=1,label='Trade Price',color='green')
ax.legend()

ax.set_title('National Best Bid and Offer for AAPL')
ax.set_ylabel('Price')
ax.set_xlabel('Time')
```
